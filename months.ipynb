```python
from utz import *
```

```python
df = concat([ read_parquet(f'{year}.pqt') for year in range(2012, 2023) ])
df
```

```python
df = df.reset_index()
df
```

```python
import plotly.express as px
```

```python
fig = px.bar(
    df[~df.station.str.contains('TOTAL')],
    x='month', y='avg weekday', color='station',
    title='Average weekday PATH ridership',
    labels={
        'station': 'Station',
        'avg weekday': 'Avg Weekday Ridership',
        'month': 'Month',
    }
).update_layout(
    title_x=0.5,
)
fig.write_image('weekdays.png', width=1200, height=600,)
fig
```
