```python
from utz import *
from numpy import dtype
from tabula import read_pdf
```

```python
cur_year = 2022
last_month = 9
```

```python
template_path = 'templates/traffic-e-zpass-usage.tabula-template.json'
with open(template_path, 'r') as f:
    rects = json.load(f)
rects
```

```python
pdf = 'data/traffic-e-zpass-usage.pdf'
tables = [
    read_pdf(
        pdf,
        pages=rect['page'],
        area=[ rect[k] for k in [ 'y1', 'x1', 'y2', 'x2', ] ],
        pandas_options={'header': None},
    )
    for rect in rects
]
tbls = []
for t in tables:
    [t] = t
    tbls.append(t)
tbls[0]
```

```python
t11, t12, t13, t14, t15, t16, t17, t18, t19, t20, t21, t22 = tbls
```

```python
def check_header_row(r, s):
    r = r.values
    assert r[0] == s, f'{r[0]} != {s}'
    assert all(isna(r[1:]))

crossings = [
  'All Crossings',
  'George Washington Bridge',
  'Lincoln Tunnel',
  'Holland Tunnel',
  'Goethals Bridge',
  'Outerbridge Crossing',
  'Bayonne Bridge',
]

def crossing_rows(t, idx, year, crossing):
    row_idx = 2 + 6*idx
    check_header_row(t.iloc[row_idx], crossing)
    counts = t.copy().iloc[(row_idx + 1):(row_idx + 6)]
    if year == cur_year:
        value_vars = R1[1:(last_month + 1)] + [R1[-1]]
    else:
        value_vars = R1[1:]
    columns = ['Type'] + value_vars
    print(f'{len(columns)} cols vs. {len(counts.columns)}: {", ".join(map(str, counts.columns.tolist()))}')
    counts.columns = columns
    counts['Year'] = year
    counts['Crossing'] = crossing
    counts = counts[['Year', 'Crossing'] + columns]
    counts = counts.iloc[:5]
    counts = counts.melt(id_vars=['Year', 'Crossing', 'Type'], value_vars=value_vars, var_name='Month', value_name='Count')
    return counts

R1 = [ '(Eastbound Traffic)', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec', 'Annual']
agg_kvs = { 
    'Type': 'Total Vehicles',
    'Crossing': 'All Crossings',
    'Month': 'Annual',
}
all_cols = ['Year'] + list(agg_kvs.keys())

def check_table(t, year):
    print(f'Year: {year}')
    check_header_row(t.iloc[0], 'TUNNELS AND BRIDGES')
    r1 = t.iloc[1]
    expected_r1 = R1.copy()
    if year >= 2018:
        expected_r1[-1] = 'Year-to-Date'
    if year == cur_year:
        actual_r1 = r1.tolist()
        expected_r1 = expected_r1[:(last_month + 1)]  # include initial "(Eastbound Traffic)" column
        if actual_r1[:len(expected_r1)] != expected_r1:
            raise RuntimeError(f"{year}: unexpected header row: {actual_r1.tolist()} doesn't start with {expected_r1}")
        columns = t.columns.tolist()
        t = t[columns[:(last_month + 1)] + [columns[-1]]]
    elif r1.tolist() != expected_r1:
        raise RuntimeError(f'{year}: unexpected header row: {r1.tolist()}')

    counts = pd.concat([
        crossing_rows(t, idx, year, crossing)
        for idx, crossing in enumerate(crossings)
    ])

    ezpass_rows = counts.Type == 'E-ZPass Usage (%)'
    ezpass = counts[ezpass_rows].copy()
    ezpass['Count'] = ezpass['Count'].str.replace('%$', '', regex=True).astype(float)
    ezpass = ezpass.rename(columns={'Count': 'E-Z Pass Percent'}).drop(columns='Type')
    counts = counts[~ezpass_rows].copy()
    counts['Count'] = counts['Count'].str.replace('[, ]', '', regex=True).astype(int)

    # Verify and remove aggregate/total rows
    filter_rows = None
    for k, v in agg_kvs.items():
        total_rows = counts[k] == v
        cols = [ col for col in all_cols if col != k ]
        totals1 = counts[ total_rows].groupby(cols).Count.sum().rename('total1').sort_index()
        totals2 = counts[~total_rows].groupby(cols).Count.sum().rename('total2').sort_index()
        if (totals1 == totals2).all():
            print(f'\tVerified sums: {k} == "{v}"')
        else:
            raise RuntimeError(f"{k}={v}: aggregation verification failure")
        if filter_rows is None:
            filter_rows = ~total_rows
        else:
            filter_rows = filter_rows & ~total_rows        

    filtered = counts[filter_rows].copy()
    return filtered, ezpass
```

```python
counts, ezpass = [
    pd.concat(dfs)
    for dfs in zip(*[
        check_table(tbl, idx + 2011)
        for idx, tbl in enumerate(tbls)
    ])
]
counts
```
