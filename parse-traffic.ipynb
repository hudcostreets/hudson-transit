```python
def check_header_row(r, s):
    r = r.values
    assert r[0] == s, f'{r[0]} != {s}'
    assert all(isna(r[1:]))

crossings = [
  'All Crossings',  # [ 2],
  'George Washington Bridge',  # [ 8],
  'Lincoln Tunnel',  # [14],
  'Holland Tunnel',  # [20],
  'Goethals Bridge',  # [26],
  'Outerbridge Crossing',  # [32],
  'Bayonne Bridge',  # [38],
]

def crossing_rows(t, idx, year, crossing):
    row_idx = 2 + 6*idx
    check_header_row(t.iloc[row_idx], crossing)
    counts = t.copy().iloc[(row_idx + 1):(row_idx + 6)]
    columns = ['Type'] + R1[1:]
    #print(f'{len(columns)} cols vs. {len(counts.columns)}: {", ".join(map(str, counts.columns.tolist()))}')
    counts.columns = columns
    counts['Year'] = year
    counts['Crossing'] = crossing
    counts = counts[['Year', 'Crossing'] + columns]
#     ezpass = counts.iloc[4]
#     totals = counts.iloc[3]
    counts = counts.iloc[:5]
    counts = counts.melt(id_vars=['Year', 'Crossing', 'Type'], value_vars=R1[1:], var_name='Month', value_name='Count')    
    return counts #, totals, ezpass

R1 = [ '(Eastbound Traffic)', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec', 'Annual']
agg_kvs = { 
    'Type': 'Total Vehicles',
    'Crossing': 'All Crossings',
    'Month': 'Annual',
}
all_cols = ['Year'] + list(agg_kvs.keys())

def check_table(t, year):
    check_header_row(t.iloc[0], 'TUNNELS AND BRIDGES')
    r1 = t.iloc[1]
    r1.tolist() == R1

    counts = pd.concat([
        crossing_rows(t, idx, year, crossing)
        for idx, crossing in enumerate(crossings)
    ])

    ezpass_rows = counts.Type == 'E-ZPass Usage (%)'
    ezpass = counts[ezpass_rows].copy()
    ezpass['Count'] = ezpass['Count'].str.replace('%$', '', regex=True).astype(float)
    ezpass = ezpass.rename(columns={'Count': 'E-Z Pass Percent'}).drop(columns='Type')
    counts = counts[~ezpass_rows].copy()
    counts['Count'] = counts['Count'].str.replace(',', '').astype(int)

    # Verify and remove aggregate/total rows
    filter_rows = None
    for k, v in agg_kvs.items():
        total_rows = counts[k] == v
        cols = [ col for col in all_cols if col != k ]
        totals1 = counts[ total_rows].groupby(cols).Count.sum().rename('total1').sort_index()
        totals2 = counts[~total_rows].groupby(cols).Count.sum().rename('total2').sort_index()
        if (totals1 == totals2).all():
            print(f'Verified sums: {k} == "{v}"')
        else:
            raise RuntimeError(f"{k}={v}: aggregation verification failure")
        if filter_rows is None:
            filter_rows = total_rows
        else:
            filter_rows = filter_rows & ~total_rows        

    filtered = counts[filter_rows].copy()
    return filtered, ezpass
```
