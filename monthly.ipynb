```python
from utz import *

from numpy import nan
import pandas as pd
from tabula import read_pdf
```

```python
pdf = '2022-11-11-PATH-Monthly-Ridership-Report.pdf'
```

```python
with open('tabula-2022-11-11-PATH-Monthly-Ridership-Report.sh', 'r') as f:
    lns = [ line.rstrip('\n') for line in f.readlines() ]

lns
```

```python
l = Series(lns)
l

areas = l.str.extract('-a (?P<area>\S+) -p (?P<page>\d+)')

def parse_area(l):
    pcs = l.split(',')
    assert len(pcs) == 4
    return [ float(pc) for pc in pcs ]

areas['area'] = areas['area'].apply(parse_area)
areas = areas[['page', 'area']]
areas
```

```python
by_page = areas.groupby('page').apply(lambda r: r.area.tolist()).to_dict()
by_page
```

```python
pages = { 
    page: read_pdf(
        pdf,
        pages=page,
        area=area,
        pandas_options={'header': None},
    ) 
    for page, area in by_page.items()
}
pages
```

```python
cols1 = [ 'station', 'total', 'avg weekday', 'avg sat', 'avg sun', 'avg holiday' ]
cols2 = [ 'station', 'avg daily', 'total weekday', 'total sat', 'total sun', 'total holiday' ]
```

```python
avgs = pd.concat([
    df.assign(date=to_dt(f'2022-{page}'))
    for page, dfs in pages.items()
    for df in dfs[:2]
])
avgs.columns = cols1 + ['month']
avgs = avgs.assign(**{
    k: avgs[k].astype(str).str.replace(',', '').astype(int)
    for k in cols1[1:]
})
avgs
```

```python
avgs.isna().sum(axis=0)
```

```python
avgs.dtypes
```

```python
sums = pd.concat([
    df.assign(date=to_dt(f'2022-{page}'))
    for page, dfs in pages.items()
    for df in dfs[2:]
])
sums.columns = cols2 + ['month']
sums = sums.assign(**{
    k: sums[k].astype(str).str.replace(',', '').astype(int)
    for k in cols2[1:]
})
sums
```

```python
sums.isna().sum(axis=0)
```

```python
sums.dtypes
```

```python
df = sxs(
    avgs.set_index(['month', 'station']),
    sums.set_index(['month', 'station']),
)
df.columns = [cols2[1]] + cols1[2:] + [cols1[1]] + cols2[2:]
df
```

```python
df.to_parquet('2022.pqt')
```
