```python
from utz import *

from numpy import nan
import pandas as pd
from tabula import read_pdf
```

```python
pdf = '2022-11-11-PATH-Monthly-Ridership-Report.pdf'
```

```python
with open('tabula-2022-11-11-PATH-Monthly-Ridership-Report.sh', 'r') as f:
    lns = [ line.rstrip('\n') for line in f.readlines() ]

lns
```

```python
l = Series(lns)
l
```

```python
areas = l.str.extract('-a (?P<area>\S+) -p (?P<page>\d+)')

def parse_area(l):
    pcs = l.split(',')
    assert len(pcs) == 4
    return [ float(pc) for pc in pcs ]

areas['area'] = areas['area'].apply(parse_area)
areas = areas[['page', 'area']]
areas
```

```python
by_page = areas.groupby('page').apply(lambda r: r.area.tolist()).to_dict()
by_page
```

```python
pages = { 
    page: read_pdf(
        pdf,
        pages=page,
        area=area,
        pandas_options={'header': None},
    ) 
    for page, area in by_page.items()
}
pages
```

```python
cols1 = [ 'station', 'month total', 'avg week', 'avg sat', 'avg sun', 'avg holiday' ]
cols2 = [ 'station', 'avg daily total', 'total week', 'total sat', 'total sun', 'total holiday' ]
avgs = pd.concat([
    df.assign(date=to_dt(f'2022-{page}'))
    for page, dfs in pages.items()
    for df in dfs[:2]
])
avgs.columns = cols1 + ['month']
avgs = avgs.assign(**{
    k: avgs[k].astype(str).str.replace(',', '').astype(int)
    for k in cols1[1:]
})
avgs
```
